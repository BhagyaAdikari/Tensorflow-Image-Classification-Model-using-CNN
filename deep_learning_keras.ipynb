```python
import tensorflow as tf
from tensorflow.keras import models, layers

# Load CIFAR-10 dataset
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.cifar10.load_data()

# Normalize the data
X_train = X_train.astype('float32') / 255.0
X_test = X_test.astype('float32') / 255.0

# Reshape y_train and y_test to have shape (50000, 1) and (10000, 1) respectively
y_train = y_train.reshape(-1, 1)
y_test = y_test.reshape(-1, 1)

# Define the model
ann = models.Sequential([
    layers.Flatten(input_shape=(32, 32, 3)),  # Input layer
    layers.Dense(3000, activation='relu'),    # Hidden layer 1
    layers.Dense(1000, activation='relu'),    # Hidden layer 2
    layers.Dense(10, activation='softmax')    # Output layer
])

# Compile the model
ann.compile(optimizer='SGD',
            loss='sparse_categorical_crossentropy',
            metrics=['accuracy'])

# Train the model
history = ann.fit(X_train, y_train, epochs=5, validation_data=(X_test, y_test))

# Evaluate the model
test_loss, test_accuracy = ann.evaluate(X_test, y_test)
print(f"Test Accuracy: {test_accuracy:.4f}")
```

Build Simple artifical neural network for image recognition

```python
import tensorflow as tf
from tensorflow.keras import models, layers

# Define the model
ann = models.Sequential([
    layers.Flatten(input_shape=(32, 32, 3)),  # Input layer
    layers.Dense(3000, activation='relu'),    # Hidden layer 1
    layers.Dense(1000, activation='relu'),    # Hidden layer 2
    layers.Dense(10, activation='softmax')    # Output layer
])

# Compile the model
ann.compile(optimizer='SGD',
            loss='sparse_categorical_crossentropy',
            metrics=['accuracy'])

# Normalize the data (if not already done)
X_train = X_train.astype('float32') / 255.0
X_test = X_test.astype('float32') / 255.0

# Train the model
history = ann.fit(X_train, y_train, epochs=5, validation_data=(X_test, y_test))

# Evaluate the model
test_loss, test_accuracy = ann.evaluate(X_test, y_test)
print(f"Test Accuracy: {test_accuracy:.4f}")
```

```python
from sklearn.metrics import confusion_matrix , classification_report
import numpy as np
y_pred = ann.predict(X_test)
y_pred_classes = [np.argmax(element) for element in y_pred]

print("Classification Report: \n", classification_report(y_test, y_pred_classes))
```
